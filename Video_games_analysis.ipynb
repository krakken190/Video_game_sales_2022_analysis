```python
top_na = df.sort_values(by="NA_Sales", ascending=False).head(10)
print("   NORTH AMERICA SALES   ")
sns.barplot(x="Name", y="NA_Sales", data=top_na)
plt.xticks(rotation=90)
plt.show()

print("   EUROPE SALES   ")
top_eu = df.sort_values(by="EU_Sales", ascending=False).head(10)
sns.barplot(x="Name", y="EU_Sales", data=top_eu)
plt.xticks(rotation=90)
plt.show()

print("   JAPAN SALES   ")
top_jp = df.sort_values(by="JP_Sales", ascending=False).head(10)
sns.barplot(x="Name", y="JP_Sales", data=top_jp)
plt.xticks(rotation=90)
plt.show()

print("   GLOBALSALES   ")
top_global = df.sort_values(by="Global_Sales", ascending=False).head(10)
sns.barplot(x="Name", y="Global_Sales", data=top_global)
plt.xticks(rotation=90)
plt.show()

print("   OTHERSALES   ")
top_other = df.sort_values(by="Other_Sales", ascending=False).head(10)
sns.barplot(x="Name", y="Other_Sales", data=top_other,)
plt.xticks(rotation=90)
plt.show()
```

```python
yearly_sales = df.groupby("Year_of_Release")[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales", "Global_Sales"]].sum().reset_index()

sns.set_style("whitegrid")
sns.lineplot(x="Year_of_Release", y="value", hue="variable", data=pd.melt(yearly_sales, 
id_vars="Year_of_Release", var_name="variable", value_name="value"))
plt.title("Trend of Sales over the Years")
plt.xlabel("Year of Release")
plt.ylabel("Total Sales (in millions)")
plt.show()
```

```python
top_dev_sales = df.groupby("Developer")["Global_Sales"].sum().reset_index().sort_values(by="Global_Sales", ascending=False).head(10)
top_dev_sales
```

```python
df["User_Score"] = pd.to_numeric(df["User_Score"], errors="coerce")
developer_stats = df.groupby("Developer")[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales", 
"Global_Sales", "Critic_Score", "User_Score"]].agg({"NA_Sales": "sum", "EU_Sales": "sum", "JP_Sales": "sum", 
"Other_Sales": "sum", "Global_Sales": "sum", "Critic_Score": "mean", "User_Score": "mean"}).reset_index()

# finding the top developers by global sales and rating score
top_sales = developer_stats.sort_values(by="Global_Sales", ascending=False).head(10)
top_rating = developer_stats.dropna(subset=["Critic_Score", "User_Score"]).sort_values(by=["Critic_Score", "User_Score"], ascending=False).head(10)

# ploting the top developers by global sales and rating score
fig, axes = plt.subplots(2, 1, figsize=(15, 12))
sns.barplot(x="Developer", y="Global_Sales", data=top_sales, ax=axes[0])
axes[0].set_title("Top 10 Developers by Global Sales")
axes[0].set_xlabel("Developer")
axes[0].set_ylabel("Global Sales (in millions)")
axes[1].bar(x=top_rating["Developer"], height=top_rating["Critic_Score"], label="Critic Score")
axes[1].bar(x=top_rating["Developer"], height=top_rating["User_Score"], label="User Score")
axes[1].set_title("Top 10 Developers by Rating Score")
axes[1].set_xlabel("Developer")
axes[1].set_ylabel("Rating Score")
axes[1].tick_params(axis="x", rotation=90, labelsize=8)
axes[1].legend()
plt.tight_layout()
```
